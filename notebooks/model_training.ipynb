# Model Training
Train a baseline model for ventilation risk prediction from sensor and weather features.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import joblib
```

```python
df = pd.read_csv('../data/raw/sensor_data.csv')
df['target'] = (df['co2_ppm'] > 1000).astype(int)
features = ['temperature_c', 'humidity_pct', 'co2_ppm', 'window_open', 'outdoor_temp_c', 'outdoor_humidity_pct']
X = df[features]
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
pred = model.predict(X_test)
print(classification_report(y_test, pred, zero_division=0))
joblib.dump(model, '../models/trained_model.pkl')
print('Saved model to ../models/trained_model.pkl')
```
